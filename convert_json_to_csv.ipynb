## business

```python
import json
business = []
with open("bakery_business.json", 'r') as f:
    business = json.loads(f.read())
f.close()
```

```python
len(business)
```

```python
att = []
for item in business:
    att.append(item['attributes'])
```

```python
att_without_none = []
for item in att:
    if item is None:
        item = {'attributes': '0'}
    else:
        item['attributes'] = '1'
    att_without_none.append(item)    
```

```python
att_without_none[155:157]
```

```python
import pandas as pd
att_df = pd.DataFrame.from_dict(att_without_none)
```

```python
for col in att_df.columns:
    print(att_df[col].dropna().head())
# BusinessParking, Ambience, GoodForMeal, music, Bestnights, DietaryRestrictions
```

```python
#att_df = att_df.drop(columns = ['BusinessParking', 'Ambience', 'GoodForMeal', 'Music', 'BestNights', 'DietaryRestrictions'])
```

```python
BP = []
import ast
for item in att:
    try:
        if item['BusinessParking'] == 'None':
            BP.append({"BusinessParking": 0})
        else:
            valid = item['BusinessParking'].replace("\'", "\"")
            BP.append(ast.literal_eval(valid))
            BP[-1]['BusinessParking'] = 1
    except:
        BP.append({"BusinessParking": 0})

BP_df = pd.DataFrame.from_dict(BP)
#BP_df
```

```python
 A = []
import ast
for item in att:
    try:
        if item['Ambience'] == 'None':
            A.append({"Ambience": 0})
        else:
            valid = item['Ambience'].replace("\'", "\"")
            A.append(ast.literal_eval(valid))
            A[-1]['Ambience'] = 1
    except:
        A.append({"Ambience": 0})

A_df = pd.DataFrame.from_dict(A)
#A_df
```

```python
M = []
import ast
for item in att:
    try:
        if item['Music'] == 'None':
            M.append({"Music": 0})
        else:
            valid = item['Music'].replace("\'", "\"")
            M.append(ast.literal_eval(valid))
            M[-1]['Music'] = 1
    except:
        M.append({"Music": 0})

M_df = pd.DataFrame.from_dict(M)
#M_df
```

```python
GFM = []
import ast
for item in att:
    try:
        if item['GoodForMeal'] == 'None':
            GFM.append({"GoodForMeal": 0})
        else:
            valid = item['GoodForMeal'].replace("\'", "\"")
            GFM.append(ast.literal_eval(valid))
            GFM[-1]['GoodForMeal'] = 1
    except:
        GFM.append({"GoodForMeal": 0})

GFM_df = pd.DataFrame.from_dict(GFM)
#GFM_df
```

```python
BN = []
import ast
for item in att:
    try:
        if item['BestNights'] == 'None':
            BN.append({"BestNights": 0})
        else:
            valid = item['BestNights'].replace("\'", "\"")
            BN.append(ast.literal_eval(valid))
            BN[-1]['BestNights'] = 1
    except:
        BN.append({"BestNights": 0})

BN_df = pd.DataFrame.from_dict(BN)
#BN_df
```

```python
DR = []
import ast
for item in att:
    try:
        if item['DietaryRestrictions'] == 'None':
            DR.append({"DietaryRestrictions": 0})
        else:
            valid = item['DietaryRestrictions'].replace("\'", "\"")
            DR.append(ast.literal_eval(valid))
            DR[-1]['DietaryRestrictions'] = 1
    except:
        DR.append({"DietaryRestrictions": 0})

DR_df = pd.DataFrame.from_dict(DR)
#DR_df
```

```python
hours = []
for item in business:
    if item['hours'] is not None:
        hours.append(item['hours'])
        hours[-1]['hours'] = 1
    else:
        hours.append({"hours": 0})
```

```python
hours_df = pd.DataFrame.from_dict(hours)
```

```python
hours_df
```

```python
import pandas as pd
bdf = pd.DataFrame.from_dict(business)
```

```python
bdf = bdf.drop(columns = ['attributes', 'hours'])
att_df = att_df.drop(columns = ['BusinessParking', 'Ambience', 'GoodForMeal', 'Music', 'BestNights', 'DietaryRestrictions'])
```

```python
business_df = pd.concat([bdf, hours_df, att_df, BP_df, A_df, M_df, GFM_df, BN_df, DR_df], axis=1)
```
